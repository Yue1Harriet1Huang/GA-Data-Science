###Read in data

```python
import pandas as pd
import numpy as np
import psycopg2
```

```python
customers = pd.read_csv('https://raw.githubusercontent.com/TeachingDataScience/data-science-course/forstudentviewing/16_databases/data/customers.csv', header=None, index_col=0)
orders = pd.read_csv('https://raw.githubusercontent.com/TeachingDataScience/data-science-course/forstudentviewing/16_databases/data/orders.csv', header=None, index_col=0)
```

#####Find first 5 customers in region 2

```python
header_row = ['firstname','lastname','address',]

customers.columns = header_row
```

```python
customers.ix[(customers[12] == 2)][:5]
```

#####Count the number of customers by State

```python
pd.DataFrame({'count' : customers.groupby( [8] ).size()})
```

#####Number of customers that have a first name that start with 'Z'

```python
len(customers.ix[(customers[2].map(lambda x: x[:1]) == "Z")])
```

#####Mean, median, min, max, and standard deviation of age per gender

```python
grouped = customers.groupby(19)

grouped[17].agg([np.mean, np.median, np.min, np.max, np.std])
```

#####Count the number of orders by age of customer

```python
cust_orders = pd.merge(orders, customers, left_on=2, right_on=7, how='left')
```

```python
cust_orders.head()
```

```python
pd.DataFrame({'count' : cust_orders.groupby( [17] ).size()})
```
