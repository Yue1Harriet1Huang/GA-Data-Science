Casual Model: My OLS regression for casual users includes the independent varibles workingday, atemp, hum, and windspeed. I also included polynomial terms for humidity and temperature.

Registered Model: I used the same independent variables in the OLS regression for registered users.

The p-values for all of the coefficients in both models were almost all statistically signficant at 0.00, except for humidity for the casual users. This highlights the general trend that it seems like the weather related variables have a greater effect on registered users, while holiday/weekday-related variables impact casual users more.

This makes sense because we might expect casual users are tourists or are looking for liesure use of the bikes, and thus are less sensitive to the weather changes because the want to use the bikes when they have the opportunity (i.e. are visiting), whereas registered users may be locals who have alternate means of commuting when the weather is bad. If casual users are liesurely users and registered users are commute users, then that also explains why casual users would be more sensitive to the day of the week, as they are more likely to use the bikes on the weekends than during the week. On the other hand, registered users are probably a mix of people using them to commute and on the weekends, and thus the difference by working day is still significant but to a lesser degree.

### Explore difference in casual and registered renters over time

```python
from datetime import datetime,date

days['dteday_c'] = days['dteday'].map(lambda x: datetime.strptime(str(x), '%Y-%m-%d'))
days['dteday_c'] = days['dteday_c'].map(lambda x: datetime.date(x))
```

```python
fig = plt.figure(figsize=(12,5))
ax = fig.add_subplot(1,1,1)

casLine = ax.plot(days['dteday_c'], days['casual'], color='m', label='Casual Users')
regLine = ax.plot(days['dteday_c'], days['registered'], color='b', label='Registered Users')

ax.legend(loc='best')
ax.set_title('Bike Users by Date')
ax.set_xlabel("Date")
ax.set_ylabel("Number of Users")
```

Looks like there are days when registered renters drops drastically. It could be on holidays, when perhaps locals are out of town. Additionally, there is a clear relationship by season, with the highest renters in summer, and increasing number through the first half of the year, and decreasing the second half of the year.

### Check for multicollinearity and correlation with independent variables

```python
from numpy import corrcoef, sum, log, arange
from numpy.random import rand
from pylab import pcolor, show, colorbar, xticks, yticks

corrmatrix = corrcoef([days['holiday'],days['workingday'],days['weathersit'],days['temp'],days['atemp'],days['hum'],days['windspeed'],days['casual'],days['registered']])

fig, ax = plt.subplots()
corr_heatmap = ax.pcolor(corrmatrix)
colorbar(corr_heatmap)
ax.set_yticks(arange(0.5,9.5))
ax.set_xticks(arange(0,9.5))
ax.set_xticklabels(['holiday','workingday','weathersit','temp','atemp','hum','windspeed','casual','registered'], rotation=45)
ax.set_yticklabels(['holiday','workingday','weathersit','temp','atemp','hum','windspeed','casual','registered'])
```

This is not the best way to think about how to create the best linear regression model, since some of the independent variables may not be correlated with the dependent variables on their own, but may show statistically significant relationships after bringing in other variables. Additionally, they may not be correlated but including them in the model may result in other variables being correlated. However, this does give a superficial look at what indepenedent variables we should keep in mind when building the model.

Working day - strong negative correlation with the count of casual users, and moderately positive correlation with registered users. 

Weather situation - looks like strong negative correlation with both casual and registered.

Holiday - stronger negative correlation with registered than with casual (which we could somewhat see in the bar graphs).

Humidity and windspeed - moderately negative correlation with both, but seems slightly more negative for registered users.

Temperature - as expected, a strong positive correlation with both.

Windspeed, humidity, and weather situation all seem to be correlated (positive or negative) with each other. Also, working day and holiday are (not surprisingly) negatively correlated. This could be an argument for not including all of those variables in the model to avoid redundant variables.

### Fitting a model

```python
c_model, r_model = lm.LinearRegression(), lm.LinearRegression()

c_vars = ['holiday','atemp', 'hum', 'workingday', 'weathersit']
r_vars = ['atemp', 'hum','workingday', 'weathersit']

c_model.fit(days[c_vars], days['casual'])
print 'Casual users - R-squared:'
print c_model.score(days[c_vars], days['casual'])

r_model.fit(days[r_vars], days['registered'])
print '\nRegistered users - R-squared:'
print r_model.score(days[r_vars], days['registered'])
```

Adding working day made a big difference for casual users, which makes sense. Casual users are likely doing more liesurely riding and are more likely to do so on the weekends vs. registered users who are using bikes for commute and liesure, and are thus less sensitive to whether something is a working day.

Removing holiday from registered didn't make a difference to the R-squared, so I removed it to keep the model simpler. Perhaps the effect from holiday seen in the graphs is taken care of by working day, which has a signficant impact.

This method allows us to look at the R-squared, but it is difficult to say that something is a good model without also considering the P-values of the coefficients. Thus I move to statsmodel to get a better picture of the whole model.

```python
import statsmodels.formula.api as smf

print 'Casual Users'
cm = smf.ols('casual ~ atemp + I(atemp**2) + hum + I(hum**2) + workingday + windspeed + windspeed', data=days)
cm_results = cm.fit()
print cm_results.summary()

print '\n \nRegistered Users'
rm = smf.ols('registered ~ atemp + I(atemp**2) + hum + I(hum**2) + workingday + windspeed', data=days)
rm_results = rm.fit()
print rm_results.summary()
```

Statsmodel helps me to revise some of the independent variables. Instead of weather situation, I included windspeed because the two are multicollinear and it did not appear that we needed both in the model, and windspeed has a more signficant effect on the model. I also included the polynomial terms for humidity and temperature given the scatter plots. 

One interesting difference is the effect of the weather-related variables on registered users. Although they are all statistically signficant for both registered and casual users, the coefficients are much larger for registered users. Perhaps casual users are less sensitive to weather conditions because they are tourists who are using the bikes whenever they need to, but registered users are locals who just use subways or other methods of transportation when the weather is bad.

Also, holiday was not statistically signficant for registered users, while it was for casual users. The coefficient was negative when included in the model, which means that on average, the marginal effect of a holiday is a decrease in casual users, but this is after taking into account workingday. If we remove working day, the marginal effect of holiday is an increase in casual users. I chose to remove it from the model due to the multicollinearity between the two variables and its limited change in the R-squared.
