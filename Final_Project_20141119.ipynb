##### What types of noise complaints are the most frequently reported?

I start by looking at the various types of complaints that are being reported at the most general level, which is by complaint type. 

```python
def counter(seq):
    seq_dict = {}
    for n in seq:
        if n in seq_dict:
            seq_dict[n] += 1
        else:
            seq_dict[n] = 1
    return seq_dict
```

```python
type_desc = noise['type']
type_dict = counter(type_desc)
type_count_data = pd.DataFrame(type_dict.items(), columns = ['type','count'])
type_count_data = type_count_data.sort(['count'], ascending=False)

print type_count_data

type_hist = type_count_data.plot(kind='bar',title='Frequency of Each Type of Noise Complaint: 3/21/2013 - 9/20/2014', legend=False, figsize=(10,5))
type_hist.set_xlabel("Complaint Type")
type_hist.set_ylabel("Count of Complaints")
```

In looking at a bar graph of the frequency of each type of noise complaint, the ambigious "Noise" is found to be the most frequent, which is not surprising since it appears to be a catch-all. This led me to further look into what specific types of noise complaints are categorized under the general "Noise" type.

```python
typenoise = noise.ix[(noise['type'] == 'Noise')]

typenoise_count_data = pd.DataFrame({'count' : typenoise.groupby( [ "type", "descriptor"] ).size()}).reset_index()
typenoise_count_data = typenoise_count_data.sort(['type','count'],ascending=False)

typenoise_count_data
```

As expected, there is a wide range of complaints that fell under the general "Noise." The results are not that interesting as they cover various different types of complaints, with the most common being after-hour construction and barking dogs.

Next, I look at the frequency of a more granular classification of noise complaint using the descriptor for all noise types.

```python
desc_detail = pd.DataFrame({'count' : noise.groupby( [ "type", "descriptor"] ).size()}).reset_index()
desc_detail = desc_detail.sort(['type','count'],ascending=False)

desc_detail
```

##### Do complaints look different on particular holidays?

Next, I thought I would explore whether or not the number of complaints was higher than the average on days you may expect them to be, such as particular holidays like the 4th of July, Halloween, Cinco de Mayo, St. Patrick's Day, etc. Of the ones listed here, only 4th of July and Cinco de Mayo occurred in the last 6 months, so I decided to focus on those.

```python
from datetime import datetime,date
noise['create'] = noise['create'].map(lambda x: datetime.strptime(str(x), '%m/%d/%Y %H:%M'))

noise['create_date'] = noise['create'].map(lambda x: datetime.date(x))
noise['create_time'] = noise['create'].map(lambda y: datetime.time(y))
noise['create_hour'] = noise['create'].map(lambda z: datetime.time(z).hour)
```

```python
day_0505 = noise.ix[(noise['create_date'] == date(2014, 5, 5))]
day_0704 = noise.ix[(noise['create_date'] == date(2014, 7, 4))]

hour_desc = noise['create_hour']
hour_dict = counter(hour_desc)
hour_count_data = pd.DataFrame(hour_dict.items(), columns = ['hour','count'])
hour_count_data = hour_count_data.sort(['hour'])

d0 = date(2014, 3, 21)
d1 = date(2014, 9, 20)
delta = d1 - d0
hour_count_data['avg_count'] = hour_count_data['count']/delta.days

d55_desc = day_0505['create_hour']
d55_dict = counter(d55_desc)
d55_count_data = pd.DataFrame(d55_dict.items(), columns = ['hour','count'])
d55_count_data = d55_count_data.sort(['hour'])

d74_desc = day_0704['create_hour']
d74_dict = counter(d74_desc)
d74_count_data = pd.DataFrame(d74_dict.items(), columns = ['hour','count'])
d74_count_data = d74_count_data.sort(['hour'])
```

```python
fig = plt.figure(figsize=(12,5))
ax = fig.add_subplot(1,1,1)

hour_line = ax.plot(hour_count_data['hour'], hour_count_data['avg_count'], color='m', label='Average (last 6mo)')
may5_line = ax.plot(d55_count_data['hour'], d55_count_data['count'], color='r', label='Cinco de Mayo 2014')
july4_line = ax.plot(d74_count_data['hour'], d74_count_data['count'], color='b', label='July 4th 2014')

ax.legend(loc='best')
ax.set_title('Complaints Each Hour')
ax.set_xlabel("Time")
ax.set_ylabel("Number of Complaints")

labels = ['12am', '1am', '2am', '3am', '4am', '5am', '6am', '7am', '8am', '9am', '10am', '11am', '12pm', '1pm', '2pm', '3pm', '4pm', '5pm', '6pm', '7pm', '8pm', '9pm', '10pm', '11pm']
plt.xticks(hour_count_data['hour'], labels, rotation=19)
```

Looking at a comparison between the average number of noise complaints each hour versus the number of noise complaints on days that we may expect to be extra noisy, it appears that actually there may be fewer noise complaints on those days if there is any difference at all, except for a peak of complaints around 10-11pm on July 4th. This may be a result of people expecting noise and thus complaining less.

##### Is there a relationship between the day of the week and the frequency or types of noise complaints?

The finding that what we would expect to be noisy days in fact had fewer noise complaints than average then begs the question of whether there are fewer noise complaints on weekend nights, following the hypothesis that perhaps people complain less about noise on days when they expect noise. I do the following analyses to explore noise complaints by the day of the week:

1) In the last week of the available data, is there a difference in the time of noise complaints depending on the day of the week? (Limited to Loud Noise/Music complaints.)

2) Is there a difference in the frequency of noise complaints depending on the day of the week looking at all of the data?

3) Is there a difference in the type of noise complaints depending on the day of the week?

```python
week_914 = noise.ix[ ( (noise['create_date'] >= date(2014, 9, 14)) & (noise['create_date'] < date(2014, 9, 21)) & 
                      (noise['descriptor'] == 'Loud Music/Party') ) ]

base_date = date(2001,1,1)
times = np.fromiter((matplotlib.dates.date2num(datetime.combine(base_date, d)) for d in week_914['create_time']), 
                    dtype = 'float', count = len(week_914['create_time']))
dates = matplotlib.dates.date2num(week_914['create_date'])
```

```python
fig = plt.figure(figsize=(40,5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = plt.subplot(1,2,1)

ax1.scatter(times, dates)
ax1.xaxis_date()
ax1.yaxis_date()
ax1.set_xlim(min(times),max(times))
ax1.set_ylim(min(dates)-1,max(dates)+1)

plt.xlabel("Time")
plt.ylabel("Number of Loud Noise/Music Complaints")

figure_title = "Loud Noise/Music Complaints During the Week of 9/14/14-9/20/14"

plt.text(0.5, 1.08, figure_title,
         horizontalalignment='center',
         fontsize=12,
         transform = ax2.transAxes)
```

The above plot ranging from Sunday to Saturday indicates that there are more loud noise/music complaints Sunday morning and afternoon than any other day, which makes sense since people are sleeping in and at home, more sensitive to noise, and we would expect that there would also be more unwanted music/noise since people are not at work. The biggest complaint day overall, however, looks like Saturday. There are increasing noise complaints throughout the week during the evenings and night, which goes against the previous hypothesis that people complain less when noise is expected.

```python
noise['create_weekday'] = noise['create_date'].map(lambda x: datetime.weekday(x))

wd_desc = noise['create_weekday']
wd_dict = counter(wd_desc)
wd_count_data = pd.DataFrame(wd_dict.items(), columns = ['create_weekday','count'])

wd_hist = wd_count_data.plot(kind='bar',title='Frequency of Noise Complaints by Day of the Week', legend=False, figsize=(10,5))
wd_hist.set_xlabel("Day of the Week")
wd_hist.set_ylabel("Count of Complaints")
wd_hist.set_xticklabels(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
```

```python
fig = plt.figure(figsize=(30,10))
ax = fig.add_subplot(1,1,1)
width = 0.1

tnoise = ax.bar(wd_type_detail_wide.index.values, wd_type_detail_wide['Noise'], width, color='c', label='Noise')
tst = ax.bar(wd_type_detail_wide.index.values+width, wd_type_detail_wide['Noise - Street/Sidewalk'], width, color='b', label='Street/Sidewalk')
tcom = ax.bar(wd_type_detail_wide.index.values+width*2, wd_type_detail_wide['Noise - Commercial'], width, color='g', label='Commercial')
tvcl = ax.bar(wd_type_detail_wide.index.values+width*3, wd_type_detail_wide['Noise - Vehicle'], width, color='r', label='Vehicle')
tpk = ax.bar(wd_type_detail_wide.index.values+width*4, wd_type_detail_wide['Noise - Park'], width, color='m', label='Park')
thow = ax.bar(wd_type_detail_wide.index.values+width*5, wd_type_detail_wide['Noise - House of Worship'], width, color='k', label='House of Worship')
thc = ax.bar(wd_type_detail_wide.index.values+width*6, wd_type_detail_wide['Noise - Helicopter'], width, color='y', label='Helicopter')
ttk = ax.bar(wd_type_detail_wide.index.values+width*7, wd_type_detail_wide['Collection Truck Noise'], width, color='0.75', label='Collection Truck')

ax.set_xticks(wd_type_detail_wide.index.values)
ax.set_xticklabels( ('Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday') )
ax.legend(loc='best',fontsize=18)
ax.set_title('Type of Noise Complaint by Day of the Week',fontsize=24)
ax.set_xlabel("Day of the Week",fontsize=24)
ax.set_ylabel("Number of Complaints",fontsize=24)
```

##### Do noise complaints vary by location?

Next, I wanted to explore whether there were any observable trends in noise complaints by location. I look at the difference in the frequncy of complaints by borough, and then bring in external data to map the zipcodes of the complaints to the specific neighborhoods in New York.

```python
borough_clean = pd.DataFrame(noise)
borough_clean = noise.ix[ ( noise['borough']<>'Unspecified' ) ]

borough_desc = borough_clean['borough']
borough_dict = counter(borough_desc)
borough_count_data = pd.DataFrame(borough_dict.items(), columns = ['borough','count'])
borough_count_data = borough_count_data.sort('borough')
```

```python
fig = plt.figure()

figure_title = 'Percent of Noise Complaints by Borough'
ax = plt.subplot(1,2,1)

plt.text(0.5, 1, figure_title, horizontalalignment='center', fontsize=14, transform=ax.transAxes)
plt.pie(borough_count_data['count'], 
        labels=borough_count_data['borough'], 
        colors=['g','c','r','m','y'], 
        autopct='%1.1f%%', 
        pctdistance=0.75,
        shadow=True, 
        startangle=90,
        labeldistance=1.1)
ax.axis('equal')
```

Although Manhattan had nearly half of all of the noise complaints, this does not reflect the total population. I brought in 2013 population estimates from the 2010 Census to see how these numbers compared per capita. 

```python
pop_dict = {}
pop_dict['BRONX']=1418733
pop_dict['BROOKLYN']=2592149
pop_dict['MANHATTAN']=1626159
pop_dict['QUEENS']=2296175
pop_dict['STATEN ISLAND']=472621
```

```python
percap_dict = {k: int(borough_dict[k])/(pop_dict[k]/1000) for k in borough_dict}
percap_count_data = pd.DataFrame(percap_dict.items(), columns = ['borough','complaints_per_1000'])
percap_count_data
```

Manhattan has nearly three times as many noise complaints per capita as compared to some of the other boroughs.
