```python
import numpy
import pandas

header_row = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num',]
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data'
heart_data = pandas.read_csv(url, header=0, na_values='?')
heart_data.columns = header_row

#1. two different ways to count rows by gender
print len(heart_data[heart_data['sex']==1])
print len(heart_data[heart_data['sex']==0])

print heart_data['sex'].value_counts()

#2. how could you do the same using SAC?
print heart_data.groupby('sex').count()

#3. compare all other values for each gender
print heart_data.groupby('sex').describe()

...

mean thal for males is higher than that for females 
may want to fill in mean by sex

...

#4. two ways to fill in missing values for thal
heart_data['thal'] = heart_data.groupby('sex').transform(lambda x: x.fillna(x.mean()))
print heart_data['thal']

heart_data['thal'].ffill()

heart_data['thal'].bfill()

heart_data['thal'].fillna(int(heart_data['thal'].mean()))
```
