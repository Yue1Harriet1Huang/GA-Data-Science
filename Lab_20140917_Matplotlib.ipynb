```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Plotting with Matplotlib

```python
X = np.linspace(-np.pi,np.pi,256, endpoint=True)
C, S = np.cos(X), np.sin(X)
```

```python
plt.plot(X,C)
plt.plot(X,S)
```

```python
plt.figure(figsize=(8,3)) #gives entire figure size, for as many plots as printing

plt.subplot(1,2,1) #if plotting multiple, identifies space and location
plt.plot(X,C)

plt.subplot(1,2,2) 
plt.plot(X,S)

plt.xlim(-4,4)
```

```python
fig = plt.figure()
plt.axes([1,1,1,1])
plt.axes([2,1,1,1])
plt.axes([3,1,1,1])

#left, bottom, width, height
#a different method to create subplots within a figure but anywhere
#no longer limited to a grid
#everything relative to the figure size e.g. plt.figure(figsize=(1,10))
```

```python
x = np.arange(50)
y = np.random.randn(50) #randn gives random z-scores
plt.scatter(x,y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(), 'k', label='one')
ax.plot(np.random.randn(1000).cumsum(), 'k-', label='two')
ax.plot(np.random.randn(1000).cumsum(), 'k--', label='three')
ax.set_title('Three random lines')
ax.legend(loc='best')
```
